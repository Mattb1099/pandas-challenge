```python
import pandas as pd
import numpy as np
```

```python
# File to Load (Remember to Change These)
school_data_to_load = "schools_complete.csv"
student_data_to_load = "students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
passing_math_count = school_data_complete[(school_data_complete["math_score"] > 69)].count()["student_name"]
passing_read_count = school_data_complete[(school_data_complete["reading_score"] > 69)].count()["student_name"]
passing_overall_count = school_data_complete[(school_data_complete["reading_score"] > 69) & (school_data_complete["math_score"] > 69)].count()["student_name"]
percent_math_pass = (passing_math_count/Total_students) * 100
percent_read_pass = (passing_read_count/Total_students) * 100
percent_overall_pass = (passing_overall_count/Total_students) * 100
# print(passing_math_count)
# print(passing_read_count)
# print(percent_math_pass)
# print(percent_read_pass)
# print(percent_overall_pass)
```
